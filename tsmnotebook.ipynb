```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('train.csv.crdownload')

df.head()
```

tp - The total accumulated precipitation (in mm) over a given time period. This includes all forms 
of precipitation, such as rain, snow, hail, etc.

u10 - The east-west wind component at 10 meters above the surface. Positive values indicate wind blowing eastward, and negative values indicate wind blowing westward.

v10 - The north-south wind component at 10 meters above the surface. Positive values indicate wind blowing northward, and negative values indicate wind blowing southward.

sp - The atmospheric pressure at the Earth's surface, measured at the location of the observation.

u100 - The east-west wind component at 100 meters above the surface.

v100 - The north-south wind component at 100 meters above the surface.

tcc - The fraction of the sky covered by clouds, measured as a percentage (0-100%).

ptype - A numeric representation of the type of precipitation (e.g., rain, snow, etc.). This can vary depending on the dataset, but it usually uses integer values to categorize different precipitation types.

t2m - The air temperature at 2 meters above the surface.

```python
# Plot the histogram of the log1p-transformed 'ptype' (without modifying the original 'ptype' column)
plt.figure(figsize=(10, 6))
plt.hist(np.log1p(df['tp']), bins=30, edgecolor='black')  # Apply log1p for visualization
plt.title('Log1p Transformation of tp Distribution')
plt.xlabel('Log1p(tp)')
plt.ylabel('Frequency')
plt.show()
```

```python
df['tp_sqrt'] = np.sqrt(df['tp'])
df['tp_sqrt'].hist(bins=30, edgecolor='black')
plt.title('Square Root Transformation of tp')
plt.show()
```

tp and ptype are highly imbalanced features, squareroot transform of tp seems to have helped somewhat but still remains imbalanced, one way of reducing imbalance in ptype migth be consolidating categories or using under/over sampling
